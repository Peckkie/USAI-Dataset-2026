```python
import ast

# Mapping จาก GroupAB label -> AB code
disease_map = {
    'Mild fatty liver':           'AB01',
    'Moderate fatty liver':       'AB02',
    'Severe fatty liver':         'AB03',
    'Cirrhosis':                  'AB04',
    'PDF 1':                      'AB05',
    'PDF 2':                      'AB06',
    'PDF 3':                      'AB07',
    'Single Mass':                'AB081',  # LiverMass
    'BileDuct_Common bile duct':  'AB082',
    'BileDuct_Right lobe':        'AB082',
    'BileDuct_Left lobe':         'AB082',
    'Gallstone':                  'AB09',
    'Renal cyst':                 'AB10',
    'Kidney_Parenchymal change':  'AB11',
    'Parenchymal change':         'AB11',
    'Renal stone':                'AB11',
}

target_diseases = set(disease_map.keys())

def parse_group(val):
    try:
        return ast.literal_eval(val)  # แปลง string list -> list จริง
    except:
        return [val]

def has_target_disease(val):
    diseases = parse_group(val)
    return any(d in target_diseases for d in diseases)

# กรองเฉพาะแถวที่มีโรคที่สนใจ
df_disease = df_filtered[df_filtered['GroupAB'].apply(has_target_disease)].copy()

print(f'ก่อนกรอง: {df_filtered.shape[0]:,} rows')
print(f'หลังกรอง: {df_disease.shape[0]:,} rows')
```

```python
def map_group_to_ab(val):
    diseases = parse_group(val)
    ab_codes = list(dict.fromkeys(  # เอา unique code รักษาลำดับ
        disease_map[d] for d in diseases if d in disease_map
    ))
    return ab_codes if len(ab_codes) > 1 else ab_codes[0] if ab_codes else None

df_disease['AB_code'] = df_disease['GroupAB'].apply(map_group_to_ab)

# ตรวจสอบผลลัพธ์
print(df_disease[['GroupAB', 'AB_code']].head(10))
print(f'\nตัวอย่าง AB_code unique:\n{df_disease["AB_code"].value_counts().head(10)}')
```
