```python
import pandas as pd
filtered_15AB = './csv/USAI_Doctor-all_2013-2023_filtered_15AB_107271.csv'

filtered_15AB = pd.read_csv (filtered_15AB)
#ใช้คอลัมน์นี้ --> img_path เป็น primary key
print(filtered_15AB .shape)
filtered_15AB.head(10)
```

```python
usai10k = './csv/fv_usai10k_all_unlearn_mlt_model.csv'  # version ลบ field test แล้ว 
fv_usai10k_all = pd.read_csv (usai10k)
fv_usai10k_all = fv_usai10k_all.loc[:, ~fv_usai10k_all.columns.str.contains('^Unnamed')]
print(fv_usai10k_all.shape)

#ใช้คอลัมน์นี้ --> path เป็น primary key
```

```python
fv_usai10k_sample = pd.read_csv(usai10k, nrows=5)
fv_usai10k_sample
```

```python
# ===================================================
# JOIN ตาราง filtered_15AB กับ fv_usai10k_all
# อิงตาม filtered_15AB โดยใช้ img_path = path
# ===================================================

# เลือกเฉพาะคอลัมน์ path + x1 ถึง x2048
x_cols = [f'x{i}' for i in range(1, 2049)]
fv_subset = fv_usai10k_all[['path'] + x_cols]

# JOIN โดยใช้ img_path (filtered_15AB) = path (fv_usai10k_all)
merged = filtered_15AB.merge(
    fv_subset,
    left_on='img_path',
    right_on='path',
    how='left'  # อิงตาม filtered_15AB
)

print(f'filtered_15AB shape : {filtered_15AB.shape}')
print(f'merged shape        : {merged.shape}')
print(f'match ได้           : {merged["path"].notna().sum():,} rows')
print(f'ไม่ match           : {merged["path"].isna().sum():,} rows')
```

```python
merged
```

```python
# ===================================================
# กรองเอาแค่แถวที่ match ครบ (ไม่มี NaN ใน x columns)
# ===================================================
merged_clean = merged.dropna(subset=['x1']).copy()

print(f'ก่อนกรอง : {merged.shape[0]:,} rows')
print(f'หลังกรอง : {merged_clean.shape[0]:,} rows')
print(f'หายไป    : {merged.shape[0] - merged_clean.shape[0]:,} rows')
```

```python
doctor_map = {
    844:  'พญ.สุภัชซา ดีวังพล',
    465:  'รศ.พญ.ปรารถนา เชาวน์ชื่น',
    401:  'พญ.จิตราภรณ์ วงศ์วิวัฒน์ไชย',
    440:  'พญ.ฐิติมา อนุกูลอนันตชัย',
    8:    'รศ.นพ.ณรงค์ศักดิ์ ขันตีแก้ว (มหาวิทยาลัยขอนแก่น)',
    2219: 'พญ. ลลิตา ทับคำภา',
    2952: 'พญ.ภัสษญา ศรีภา',
    12:  'อ.นิตยา',
    349:  'อ.วัลลภ',
    10:  'อ.ผลิญ',
    465:  'อ.ปรารถนา',
    1816: 'อ.วรินทร',
    844:  'พญ.สุภัชชา',
}

# กรองเฉพาะแพทย์ที่ต้องการ
df_filtered_ = merged_clean[merged_clean['Doctor'].isin(doctor_map.keys())].copy()

# Map ชื่อแพทย์
df_filtered_['doctor_name'] = merged_clean['Doctor'].map(doctor_map)

# สรุปจำนวน
summary = df_filtered_.groupby(['Doctor', 'doctor_name']).size().reset_index(name='count')
summary = summary.sort_values('count', ascending=False)
summary
```

```python
summary = df_filtered_.groupby('Doctor').agg(
    NumberOfCase = ('CASE_doctor', 'nunique'),  # นับ case ไม่ซ้ำ
    NumberOfImg = ('originalImage', 'count')   # นับภาพทั้งหมด
).reset_index()

summary['%Case'] = (summary['NumberOfCase'] / summary['NumberOfCase'].sum() * 100).round(2)
summary = summary.sort_values('NumberOfCase', ascending=False)
print(summary.to_string(index=False))
```
